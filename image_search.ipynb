```python
# Find defect ratio in each square of different radius.
dict_ratio = {}
for j in list_wafer:
    data_temp = data_corr[data_corr['SCRIBE_ID']==j]
    data_temp['defect'] = data_temp['SOFTBIN'].apply(lambda x: 0 if x>1 else 1)
    list_ratio = []
    # Transfer x,y center into the middle die's right corner
    int_center_x = round((data_temp['DIE_X'].max() + data_temp['DIE_X'].min())/2)
    int_center_y = round((data_temp['DIE_Y'].max() + data_temp['DIE_Y'].min())/2)
    print('int_center_x:',int_center_x,'int_center_y:',int_center_y)
    data_temp['DIE_X_2'] = data_temp['DIE_X'] - int_center_x
    data_temp['DIE_Y_2'] = data_temp['DIE_Y'] - int_center_y
    data_temp['distance'] = data_temp['DIE_X_2']**2 + data_temp['DIE_Y_2']**2
    for i in range(1,data_temp['DIE_X'].max()):
        data_square = data_temp[(data_temp['distance'] <= i**2) & (data_temp['distance'] > (i-1)**2)]
        float_ratio = data_square['defect'].sum()/data_square.shape[0]
        list_ratio.append(float_ratio)
    dict_ratio[j] = list_ratio
```

```python
# Find defect ratio in each column.
dict_ratio = {}
for j in list_wafer:
    data_temp = data_corr[data_corr['SCRIBE_ID']==j]
    data_temp['defect'] = data_temp['SOFTBIN'].apply(lambda x: 0 if x>1 else 1)
    list_ratio = []
    for i in range(data_temp['DIE_X'].min(),data_temp['DIE_X'].max()+1):
        data_square = data_temp[data_temp['DIE_X']==i]
        float_ratio = 1 - data_square['defect'].sum()/data_square.shape[0]
        list_ratio.append(float_ratio)
    dict_ratio[j] = list_ratio
```

```python
data_ratio = pandas.DataFrame(dict_ratio)
data_result = data_ratio.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = plt.subplots(figsize=(20,20))
ax = sns.heatmap(data_result, cmap='Blues', annot=True, vmax=1, square=1)


```
